```python
def Gaussian(index):
    import numpy as np
    
    # Gaussian 1
    mean1 = [0, 0]
    cov1 = [[1, 0], [0, 1]]
    
    # Gaussian 2
    mean2 = [3, 0]
    cov2 = [[1, 0], [0, 1]]
    
    # Gaussian 3
    mean3 = [0, 3]
    cov3 = [[1, 0], [0, 1]]
    
    if index == 0:
        x, y = np.random.multivariate_normal(mean1, cov1, 1).T
    elif index == 1:
        x, y = np.random.multivariate_normal(mean2, cov2, 1).T
    elif index == 2:
        x, y = np.random.multivariate_normal(mean3, cov3, 1).T
    
    return x[0], y[0]



```

```python
def update_ci(x_i,miu):
    import numpy as np
    norm_square = []
    for i in range(0,len(miu)):
        value = np.linalg.norm(x_i-miu[i])**2
        norm_square.append(value)
    k = np.argmin(norm_square)
    return k
    
```

```python
def update_ci_list(x_vector,miu_K):
    ci_list = []
    for i in range(0,x_vector.shape[0]):
        value = update_ci(x_vector[i],miu_K)
        ci_list.append(value)
    return ci_list
```

```python
def update_miu(x_vector, updated_ci):
    import numpy as np
    n_k_list = []
    for i in range(0,len(np.unique(updated_ci))):
        count = 0
        for s in range(0,x_vector.shape[0]):
            if updated_ci[s] == np.unique(updated_ci)[i]:
                count += 1
        n_k_list.append(count)
        
    
    updated_miu_list = []
    for i in range(0,len(np.unique(updated_ci))):
        update_miu = 0
        for j in range(0,x_vector.shape[0]):
            if updated_ci[j] == np.unique(updated_ci)[i]:
                update_miu += x_vector[j]
        value = update_miu/n_k_list[i]
        updated_miu_list.append(value)
    
    return updated_miu_list
```

```python
def kmean_ci(x_vector, T, initial_miu_K):
    ci = []
    miu = []
    for i in range(0,T):
        if i == 0:
            ci_value = update_ci_list(x_vector,initial_miu_K)
            miu_value = update_miu(x_vector,  ci_value)
            ci.append(ci_value)
            miu.append(miu_value)
    
        else:
            ci_value = update_ci_list(x_vector,miu_value)
            miu_value = update_miu(x_vector,  ci_value)
            ci.append(ci_value)
            miu.append(miu_value)
    
    return ci
```

```python
def kmean_miu(x_vector, T, initial_miu_K):
    ci = []
    miu = []
    for i in range(0,T):
        if i == 0:
            ci_value = update_ci_list(x_vector,initial_miu_K)
            miu_value = update_miu(x_vector,  ci_value)
            ci.append(ci_value)
            miu.append(miu_value)
    
        else:
            ci_value = update_ci_list(x_vector,miu_value)
            miu_value = update_miu(x_vector,  ci_value)
            ci.append(ci_value)
            miu.append(miu_value)
    
    return miu
```
